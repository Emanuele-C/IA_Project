# Reading Dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Reading the starting dataset sappointment_data.csv
data= pd.read_csv("/appointment_data.csv")
data
```

```python
X = data.drop(['No-show'], axis = 1)
y = data['No-show']

X
```

```python
y
```

```python
# Splitting dataset (train,test)
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state = 1)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# splitting dataset in train.csv and test.csv. So, join X and y.
train = y_train.to_frame().join(X_train)
test = y_test.to_frame().join(X_test)
```

```python
train
```

```python
test
```

```python
# Creation of csv.
train.to_csv("train.csv", index = False)
test.to_csv("test.csv", index = False)
```
